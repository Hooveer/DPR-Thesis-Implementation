```python
#Import Necessary Libraries
import pandas as pd
import numpy as np
import nltk
#nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter
import re
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load Dataset
df = pd.read_csv('raw_dataset - Sheet1.csv')
```

```python
df
```

# Data Preprocessing

```python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    
    return text

# Feature extraction function
def extract_features(text):
    # Tokenize words and sentences
    words = word_tokenize(text)
    sentences = sent_tokenize(text)
    
    # Exclude special characters from word count but retain them in the text
    # Count only alphabetic tokens as words
    word_count = len([word for word in words if word.isalpha()])
    
    # Calculate phraseology features
    avg_word_length = np.mean([len(word) for word in words if word.isalpha()]) if word_count > 0 else 0
    avg_sentence_length = word_count / len(sentences) if len(sentences) > 0 else 0
    
    # Calculate punctuation features
    punctuation_count = sum(1 for char in text if char in ['.', ',', ':', ';', '?', '!', '-', '\'', '\"'])
    punctuation_ratio = punctuation_count / word_count if word_count > 0 else 0
    
    # Calculate linguistic diversity features
    vocab_size = len(set([word for word in words if word.isalpha()]))
    vocab_ratio = vocab_size / word_count if word_count > 0 else 0
    
    return {
        'word_count': word_count,
        'sentence_count': len(sentences),
        'avg_word_length': avg_word_length,
        'avg_sentence_length': avg_sentence_length,
        'punctuation_count': punctuation_count,
        'punctuation_ratio': punctuation_ratio,
        'vocab_size': vocab_size,
        'vocab_ratio': vocab_ratio
    }
```

```python
# Apply preprocessing and feature extraction to each text sample
df['preprocessed_text'] = df['text'].apply(preprocess_text)
df_features = df.apply(lambda row: extract_features(row['preprocessed_text']), axis=1).apply(pd.Series)
df = pd.concat([df, df_features], axis=1)
```

```python
authors = df['author'].unique()

for author in authors:
    fig, axes = plt.subplots(nrows=len(df_features.columns), ncols=1, figsize=(10, 20))
    df_author = df[df['author'] == author]
    
    for i, column in enumerate(df_features.columns):
        df_author.boxplot(column=[column], ax=axes[i])
        axes[i].set_title(f'{column} Distribution for {author}')
    
    plt.tight_layout()
    plt.show()
```

```python
for author in authors:
    df_author = df[df['author'] == author]
    
    # Select only numeric columns
    df_author_numeric = df_author.select_dtypes(include=['float64', 'int64'])
    
    # Check if there are any numeric columns before computing correlation
    if len(df_author_numeric.columns) > 0:
        corr_matrix = df_author_numeric.corr()

        plt.figure(figsize=(10, 8))
        plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
        plt.title(f'Correlation Heatmap for {author}')
        plt.colorbar()
        tick_marks = [i for i in range(len(df_author_numeric.columns))]
        plt.xticks(tick_marks, df_author_numeric.columns, rotation=45)
        plt.yticks(tick_marks, df_author_numeric.columns)
        plt.tight_layout()
        plt.show()
    else:
        print(f"No numeric columns for author: {author}")
```

```python
for column in df_features.columns:
    plt.figure(figsize=(10, 6))
    df.groupby('author')[column].mean().plot(kind='bar')
    plt.title(f'Mean {column} by Author')
    plt.xlabel('Author')
    plt.ylabel(column)
    plt.tight_layout()
    plt.show()
```
